```python
import matplotlib.pyplot as plt

def plot(df1,df2, df3, df4, df5, df6):
    fig, ax = plt.subplots()
    fig.subplots_adjust(right=0.75)

    twin1 = ax.twinx()
    twin2 = ax.twinx()
    twin3 = ax.twinx()
    twin4 = ax.twinx()
    twin5 = ax.twinx()
    

    # Offset the right spine of twin2.  The ticks and label have already been
    # placed on the right by twinx above.
#     twin2.spines.right.set_position(("axes", 1.2))
    
#     twin1.spines['right'].set_position(('outward', 40))

    p1, = ax.plot(df3['Time (s)'], df3['Total displacement (m), Point: 12'], "r--", label="Left sensor value")
    p2, = twin1.plot(df2['Time (s)'], df2['Pressure (Pa), Point: 5'], "b--", label="Right sensor value")
    p3, = twin2.plot(df1['Time (s)'], df1['Velocity field, y component (m/s), Point: 8'], "g--", label="Left sensor value")
    p4, = twin3.plot(df4['Time (s)'], df4['Velocity magnitude (m/s), Point: 8'], "c--", label="Right sensor value")
    p5, = twin4.plot(df5['Time (s)'], df5['Vorticity magnitude (1/s), Point: 3'], "y--", label="Left sensor value")
    p6, = twin5.plot(df6['Time (s)'], df6['Kinetic energy density (J/m^3), Point: 12'], "m--", label="Right sensor value")
    
    ax.set_xlim(min(df1['Time (s)']), max(df1['Time (s)']))
    ax.set_ylim(min(df3['Total displacement (m), Point: 12']), max(df3['Total displacement (m), Point: 12']))
    twin1.set_ylim(min(df2['Pressure (Pa), Point: 5']), max(df2['Pressure (Pa), Point: 5']))
    twin2.set_ylim(min(df1['Velocity field, y component (m/s), Point: 8']), max(df1['Velocity field, y component (m/s), Point: 8']))
    twin3.set_ylim(min(df4['Velocity magnitude (m/s), Point: 8']), max(df4['Velocity magnitude (m/s), Point: 8']))
    twin4.set_ylim(min(df5['Vorticity magnitude (1/s), Point: 3']), max(df5['Vorticity magnitude (1/s), Point: 3']))
    twin5.set_ylim(min(df6['Kinetic energy density (J/m^3), Point: 12']), max(df6['Kinetic energy density (J/m^3), Point: 12']))
#     twin5.set_ylim(min(df1['Velocity field, y component (m/s), Point: 8']), max(df1['Velocity field, y component (m/s), Point: 8']))

    ax.set_xlabel("Time (Sec)")
    ax.set_ylabel("Oar displacement (m)")
    twin1.set_ylabel("Pressure (Pa)")
    twin2.set_ylabel("Velocity field (m/s)")
    twin3.set_ylabel("Oar velocity (m/s)")
    twin4.set_ylabel("Vorticity (1/s)")
    twin5.set_ylabel("Kinetic energy density (J/m^3)")

    ax.yaxis.label.set_color(p1.get_color())
    twin1.yaxis.label.set_color(p2.get_color())
    twin2.yaxis.label.set_color(p3.get_color())
    twin3.yaxis.label.set_color(p4.get_color())
    twin4.yaxis.label.set_color(p5.get_color())
    twin5.yaxis.label.set_color(p6.get_color())
    #twin6.yaxis.label.set_color(p3.get_color())
    twin1.spines['right'].set_position(('outward', 0))
    twin2.spines['right'].set_position(('outward', 45))
    twin3.spines['right'].set_position(('outward', 85))
    twin4.spines['right'].set_position(('outward', 115))
    twin5.spines['right'].set_position(('outward', 155))


    tkw = dict(size=4, width=1.5)
    ax.tick_params(axis='y', colors=p1.get_color(), **tkw)
    twin1.tick_params(axis='y', colors=p2.get_color(), **tkw)
    twin2.tick_params(axis='y', colors=p3.get_color(), **tkw)
    twin3.tick_params(axis='y', colors=p4.get_color(), **tkw)
    twin4.tick_params(axis='y', colors=p5.get_color(), **tkw)
    twin5.tick_params(axis='y', colors=p6.get_color(), **tkw)
    ax.tick_params(axis='x', **tkw)

    #ax.legend(handles=[p1, p2, p3, p4, p5, p6], loc='lower left')
    #ax.set_title(S+" (right sensor)")

    plt.show()
```

```python
plot(df1, df2,df3, df4, df5, df6)
```
