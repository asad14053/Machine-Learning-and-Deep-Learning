```python
# read the iris.data file into a DataFrame, renaming the 5 columns to be sepal_length_cm,sepal_width_cm,petal_length_cm,petal_width_cm, and species and display the "info" about the resulting dataframe
df3 = pd.read_csv("data/iris.data", header =None)
df3.columns
df3.rename(columns = {0:"sepal_length_cm", 1: "sepal_width_cm", 2: "petal_length_cm", 3: "petal_width_cm", 4 : "species"}, inplace =True)
df3.info()
#df3
```

```python
# read the titanic.csv dataset into a dataframe, with the index set to PassengerId

df7 = pd.read_csv("data/titanic.csv", index_col= 'PassengerId')
df7
```

```python
# drop the column Cabin, and drop all the missing values in "Embarked", then show the info
del df7['Cabin']
df8 = df7.dropna(subset=['Embarked'])
df8.info()
```

```python
# fill any missing age with the average age of those that you have, and show the info for the result
#df8.columns
df8['Age'].fillna(df8['Age'].mean(), inplace=True)
df8.info()
```

```python
# add a new column called FamilySize to the dataframe, which is the SibSp + the Parch columns + 1 (themself)
df8['FamilySize'] = df.SibSp
#print(id(df.new_col), id(df.first_name))
df8['FamilySize'] += (df.Parch+1)
df8
```

```python
# describe the numerical variables
df8.describe()
```
