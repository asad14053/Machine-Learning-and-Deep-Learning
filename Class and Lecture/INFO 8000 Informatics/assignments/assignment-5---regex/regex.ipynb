```python
#read the TS in Harry Potter 1.csv (note, this is a semicolon separated dataset)
#this dataset was retrieved from https://www.kaggle.com/datasets/gulsahdemiryurek/harry-potter-dataset

df = pd.read_csv("data/Harry Potter 1.csv", sep=';')
df
```

```python
#get all sentences that end in a question mark spoken by Ron (hint, use the str.contains method of a series)

df1 = df.query('Character == "Ron"')
index = df1.Sentence.str.contains(r'\?$')
df1[index]
```

```python
#get all of the sentences that mention both Harry and Ron
index = df.Sentence.str.contains(r"\bHarry.*\b\bRon.*\b|\bRon.*\b.*\bHarry.*\b")
df[index]
```

```python
#get all of the sentences that start with anumber

index = df.Sentence.str.contains(r"^\d.*$")
df[index]
```

```python
#get all sentences that mention a professor at the end of a sentence excluding punctuation, and extract just name of the professor mentioned (hint, concatenate the result of the extract operation with the original dataframe along axis=1) and then dropna() to eliminate the ones that had no mention
df2 = df.Sentence.str.extract(r"\bProfessor\b\s(\b\w*\b)\W*$")
df3 = pd.concat([df2,df], axis =1).dropna()
df3.rename(columns={0:"Mentioned"}, inplace= True)
df3
```

```python
#extract quotes contained within any sentence

df2 = df.Sentence.str.extract(r"\"(.*)\"")
df3 = pd.concat([df2,df], axis =1).dropna()
df3.rename(columns={0:"Quoted"}, inplace= True)
df3
```

```python
#extract the first hyphenated word in each sentence, where the first word is at least 2 characters
df2 = df.Sentence.str.extract(r"((?:\w{2,}-)+\w*)")
df3 = pd.concat([df2,df], axis =1).dropna()
df3.rename(columns={0:"Hyphenated"}, inplace= True)
df3
```

```python
#extract a list of long words (>= 10 characters) from all sentences (findall is probably the easiest thing to use) that have at least two long words.  This is easier to do as a regex followed by a subset

df01 = df.Sentence.str.extract(r"(:?\w{10,}\b)(?:.*)(:?\w{10,}\b)")
df2 = '['+df01[0] +', '+ df01[1]+']'
df3 = pd.concat([df2,df], axis =1).dropna()
df3.rename(columns={0:"LongWords"}, inplace= True)
df3
```

```python

#extract sentences where the speaker mentions themselves

a_list = []
df = pd.read_csv("data/Harry Potter 1.csv", sep=';')
#df2 = pd.DataFrame(OrderedDict{'Character': 'x','Sentence': 'y'})

for index, row in df.iterrows():
  s1 = row['Character']
  s2 = row['Sentence']
  #s3 =r'(\b'+ re.escape(s1) +r'\b)'
  #print(s3)
  f = s2.find(s1)

  #print(f>=0)
  if f >-1:
    a_list.append(True)
  else:
    a_list.append(False)

df[a_list]
```

```python
#display sentences where any word is repeated 4 or more times (note you may get a warning)
df2 = df.Sentence.str.extract(r"(\b\w{2,}\W){3,}")
#df[df.str.count(r'(^F.*)')==1]
df3 = pd.concat([df2,df], axis =1).dropna()
df3.rename(columns={0:"RepeatedWord"}, inplace= True)
df3
```
